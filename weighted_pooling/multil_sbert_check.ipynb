```python
from sentence_transformers import SentenceTransformer, models

MODEL_NAME = 'sentence-transformers/paraphrase-multilingual-mpnet-base-v2'

transformer = models.Transformer(MODEL_NAME)
pooling = models.Pooling(transformer.get_word_embedding_dimension(), pooling_mode='mean')
model_as_sbert = SentenceTransformer(modules=[transformer, pooling])
```

```python
def sbert_tokenizer(model, strs):
  st_tokenizer = model._first_module().tokenizer
  tokens_list = model._first_module().tokenize(strs)['input_ids']

  for n, token_ids in enumerate(tokens_list):
    tokens = st_tokenizer.convert_ids_to_tokens(token_ids)
    print(token_ids)
    print(tokens)

  return tokens_list
```

```python
sentence1 = ['千葉県']
sentence2 = ['千葉県<pad>の田舎']

tokens_list1 = sbert_tokenizer(model_as_sbert, sentence1)
tokens_list2 = sbert_tokenizer(model_as_sbert, sentence2)

print(model_as_sbert.tokenize(sentence1))
```

```python
sentence1_sbert_embeddings = model_as_sbert.encode(sentence1)
print(len(sentence1_sbert_embeddings[0]))
print(sentence1_sbert_embeddings)
```
