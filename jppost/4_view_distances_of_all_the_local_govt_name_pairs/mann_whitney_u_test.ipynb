```python
# '1_multil_'
# '2_sonoisa_'
# '3_tohoku_'
# '4_izumil_'

model1_name_str = '2_sonoisa_'
model2_name_str = '1_multil_20230404_021123_'

# alternative = 'two-sided'
alternative = 'less' # model1 < model2  i.e. F(u) > G(u) for model1 ~ F, model2 ~ G
# alternative = 'greater' # model1 > model2  i.e. F(u) < G(u) for model1 ~ F, model2 ~ G
```

```python
import pandas as pd


dirname_str = '0_old/'
filename_base_str = 'cossim_melted_'
chimei_types_list = [
    '00_prefectures',
    '01_shi',
    '02_ku',
    '03_gun',
    '04_cho',
    '05_son'
]
file_suffix_str = '.csv'

filepath_output_str = 'mann_whitney_u/mann_whitney_u_' + model1_name_str + model2_name_str + alternative + file_suffix_str
print(filepath_output_str)

cossim_melted1_dfs_dict = dict()
cossim_melted2_dfs_dict = dict()

for chimei_type in chimei_types_list:
    filepath1_str = dirname_str + filename_base_str + model1_name_str + chimei_type + file_suffix_str
    filepath2_str = dirname_str + filename_base_str + model2_name_str + chimei_type + file_suffix_str
    
    print(filepath1_str)
    print(filepath2_str)
    
    cossim_melted1_dfs_dict[chimei_type] = pd.read_csv(filepath1_str)
    cossim_melted2_dfs_dict[chimei_type] = pd.read_csv(filepath2_str)
    
    print(cossim_melted1_dfs_dict[chimei_type].head(3))
    print(cossim_melted1_dfs_dict[chimei_type].head(3))
    
```

```python
import scipy.stats as st


mannwhitneyu_sr = pd.Series()
for chimei_type in chimei_types_list:
    
    cossim_melted1_list = cossim_melted1_dfs_dict[chimei_type]['cos_sim'].tolist()
    cossim_melted2_list = cossim_melted2_dfs_dict[chimei_type]['cos_sim'].tolist()
    
    # print(cossim_melted1_list[:10])
    # print(cossim_melted2_list[:10])
    
    mannwhitneyu_sr[chimei_type] = st.mannwhitneyu(
        cossim_melted1_list,
        cossim_melted2_list,
        alternative=alternative
    ).pvalue

print(mannwhitneyu_sr)
```

```python
mannwhitneyu_sr.to_csv(filepath_output_str)
```
