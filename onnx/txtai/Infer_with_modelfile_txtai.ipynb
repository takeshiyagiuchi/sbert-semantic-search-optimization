```python
model_path_str = "onnxed_models/izumil_jppost-cp-e1-b512_jsts-100-cp-e3-b1_onnx-pl-qT-op12"
tokenizer_path_str = "../../fine-tuning/tuned_models/izumil_jppost-cp-e1-b512_jsts-100-cp-e1to5-b1/checkpoints/37353"
max_length_token_int = 128

print(f"{model_path_str = }")
print(f"{tokenizer_path_str = }")
print(f"{max_length_token_int = }")
```

```python
import onnx

model = onnx.load(model_path_str)
print(onnx.checker.check_model(model))
```

```python
from onnxruntime import InferenceSession, SessionOptions
from transformers import AutoTokenizer


tokenizer = AutoTokenizer.from_pretrained(tokenizer_path_str)

options = SessionOptions()
# session = InferenceSession(model, options, providers = ["CPUExecutionProvider"])
session = InferenceSession(model_path_str, options, providers = ["CPUExecutionProvider"])
```

```python
tokens_list = tokenizer(["これはテストです"], padding=True, truncation=True, max_length=max_length_token_int, return_tensors='np')
embeddings_list = session.run(None, dict(tokens_list))[0]

print(embeddings_list)
```
