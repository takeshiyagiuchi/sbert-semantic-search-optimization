```python
FILEPATH_VOCAB_NEW = 'new_vocab_tohokuw_1_list.txt'

tokens_new_str_list = list()
with open(FILEPATH_VOCAB_NEW) as fp:
    for line in fp:
        tokens_new_str_list.append(line.rstrip())

print(tokens_new_str_list)
```

```python
from transformers import AutoTokenizer, AutoModel

model_name = 'cl-tohoku/bert-base-japanese-whole-word-masking'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
tokens_old_str_list = [token for token, index in tokenizer.get_vocab().items()]

print(tokens_old_str_list[:50])
```

```python
indexes_old_list = list()
vocab_common_tuple_list = list()
vocab_differ_tuple_list = list()

for index_new_int, token in enumerate(tokens_new_str_list):
    try:
        index_old_int = tokens_old_str_list.index(token)
    except:
        index_old_int = -1
    
    if index_old_int >= 0:
        indexes_old_list.append(index_old_int)
        vocab_common_tuple_list.append((token, index_new_int))
    else:
        vocab_differ_tuple_list.append((token, index_new_int))

print(f"{vocab_common_tuple_list = }")
print(f"{vocab_differ_tuple_list = }")
```

```python
modelpath_to_save = 'tohoku-bbwwm_vocab-added'

added_tokens = tokenizer.add_tokens([token_tuple[0] for token_tuple in vocab_differ_tuple_list])
print(f"{added_tokens = }")

model.resize_token_embeddings(len(tokenizer))

model.save_pretrained(modelpath_to_save)
tokenizer.save_pretrained(modelpath_to_save)
```

```python
import os
import json

filepath_added_tokens = modelpath_to_save + "/added_tokens.json"

with open(filepath_added_tokens) as fp:
    added_vocab_dict = json.load(fp)

sorted_vocab_dict = dict(sorted(added_vocab_dict.items(), key=lambda item: item[1]))
sorted_tokens_str_list = [token_str for token_str in sorted_vocab_dict.keys()]

print(sorted_vocab_dict)
print(sorted_tokens_str_list)
```

```python
filepath_vocab_str = modelpath_to_save + "/vocab.txt"

with open(filepath_vocab_str, "a", encoding="utf-8") as fp:
    for token_str in sorted_tokens_str_list:
        fp.write(f"{token_str}\n")
```
