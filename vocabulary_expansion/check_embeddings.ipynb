```python
from sentence_transformers import SentenceTransformer, models

MODEL_PATH = './sonoisa'
# MODEL_PATH = 'sonoisa/sentence-bert-base-ja-mean-tokens-v2'

transformer = models.Transformer(MODEL_PATH)
pooling = models.Pooling(transformer.get_word_embedding_dimension())
model = SentenceTransformer(modules=[transformer, pooling])
```

```python
sentence1 = ['砺波']
sentence2 = ['砺波市']
# sentence1 = ['千葉']  # worked!
# sentence2 = ['千葉県']  # worked!

sentence1_embeddings = model.encode(sentence1)
sentence2_embeddings = model.encode(sentence2)

print(sentence1_embeddings)
print(sentence2_embeddings)
```
